```python
#Bài 2
a = input("Nhập chiều rộng khu vườn: ")
b = input("Nhập chiều dài khu vườn: ")
dien_tich = float(a)*float(b)
dien_tich_tron = 3.14*(float(a)/2)**2
if a > b:
    print("Số liệu không hợp lệ")
elif float(a) <= 0 or float(b) <= 0:
    print("Số liệu không hợp lệ")
else:
    print("Diện tích phần đất còn lại là: ", round(dien_tich - dien_tich_tron, 2))
```

```python
#Bài 5
word = input("Nhập chữ cái in hoa đầu tiên: ")
c = word.lower()
if len(c) != 1 or not c.isalpha():
    print("Vui lòng nhập lại duy nhất 1 chữ cái")
else:
    if c == 'a':
        print("Đây là trường hợp đặc biệt, vui lòng nhập chữ khác")
    elif c != 'a':
        ma_so = ord(c)
        ma_truoc = chr(ma_so - 1)
        print("Chữ thường đằng trước chữ đã nhập là", ma_truoc)
    else:
        print("Lỗi thông tin đầu vào")
```

```python
#Bài 6
import math
a = float(input("Nhập số liệu đầu tiên: "))
b = float(input("Nhập số liệu thứ 2: "))
c = float(input("Nhập số liệu thứ 3: "))
p = (a+b+c)/2
dien_tich = round(math.sqrt(p*(p-a)*(p-b)*(p-c)), 1)
if a+b <= c or b+c <= a or a+c <= b:
    print("Không phải 3 cạnh của tam giác")
else:
    print("Diện tích tam giác đó là:", dien_tich)
```
